```python
import pandas as pd
import polars as pl
from sklearn.feature_selection import SelectKBest, f_classif
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

path='https://frenzy86.s3.eu-west-2.amazonaws.com/python/data/esercizi/sonar.csv'
```

```python
df=pl.read_csv(path,has_header=False)
df
```

```python
df.columns
```

```python
X=df.drop("column_61")
y=df.select("column_61")

diz_rocks={'M':0,'R':1}

#y=y.map_rows(lambda x:0 if x=="M" else 1)

X_train, X_test, y_train, y_test = train_test_split(X, y,
                                                    test_size=0.25,
                                                    random_state=667,
                                                    )
```

```python
from sklearn.feature_selection import RFECV
import numpy as np
base_model = RandomForestClassifier(n_estimators=100, random_state=42)
y_train=np.ravel(y_train)
y_test=np.ravel(y_test)

rfe = RFECV(estimator=base_model, min_features_to_select=10, step=1)
rfe.fit(X_train, np.ravel(y_train))
```

```python
print(len(rfe.ranking_))
print(len(rfe.support_))
print(len(df.columns))
```

```python

feature_ranking = pd.DataFrame({
                                'Feature': X.columns,
                                'Selected': rfe.support_,
                                'Ranking': rfe.ranking_
                                })
feature_ranking = feature_ranking.sort_values('Ranking')
feature_ranking
```

```python
ranking=np.ravel(feature_ranking['Feature'])
test_ranking=[]
for feature in ranking:
    test_ranking.append(feature)
    X_train_selected = X_train[test_ranking]
    X_test_selected = X_test[test_ranking]

    final_model = RandomForestClassifier(n_estimators=100,
                                    random_state=42,
                                    max_depth=4
                                    )
    final_model.fit(X_train_selected, y_train)
    y_pred = final_model.predict(X_test_selected)

    accuracy = accuracy_score(y_test, y_pred)

    print(f"\nFeature selezionate:\n{len(test_ranking)}")
    print(f"\nAccuratezza del modello con le feature selezionate: {accuracy:.3f}")
```
