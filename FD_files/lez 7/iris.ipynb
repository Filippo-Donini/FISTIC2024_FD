```python
import pandas as pd
import numpy as np

df=pd.read_csv('https://frenzy86.s3.eu-west-2.amazonaws.com/fav/iris.data')
print(df.columns)
df.loc[-1] = df.columns
df.index = df.index + 1
df = df.sort_index()
df.columns = ['sepal length', 'sepal width', 'petal length', 'petal width', 'class']
df
```

```python
feature_cols = ['pregnant', 'insulin', 'bmi', 'age','glucose','bp','pedigree']
X = df.drop('class',axis=1) # Features
y = df['class']
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=16)
y_test
```

```python
from sklearn.linear_model import LogisticRegression

# instantiate the model (using the default parameters)
logreg = LogisticRegression(random_state=16)

# fit the model with data
logreg.fit(X_train, y_train)

y_pred = logreg.predict(X_test)
y_pred
```

```python
df_results=pd.DataFrame(y_test)
df_results['pred']=y_pred
df_results
```

```python
species={'Iris-versicolor':0,'Iris-setosa':1,'Iris-virginica':2}
```

```python
df_results['class']=df_results['class'].map(species)
df_results['pred']=df_results['pred'].map(species)
df_results
```

```python
df_results['results']=df_results['class']-df_results['pred']
df_results['results'].value_counts()
```

```python
import pickle

with open("model.pkl", "wb") as f:
    pickle.dump(logreg, f)
```
